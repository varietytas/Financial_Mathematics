$\textcolor{lightblue}{\textbf{Q1. (2 points)}}$ 

Consider the linear SDE  

$$
dX_t = a(t)X_t\,dt + b(t)X_t\,dW_t, \quad X_0 > 0.
$$

(a) Derive the explicit solution using Itô’s formula.  

(b) Show that  

$$
X_t = X_0 \exp\!\left(\int_0^t (a(s) - \tfrac{1}{2}b(s)^2)\,ds + \int_0^t b(s)\,dW_s\right).
$$  

(c) Explain in words why $X_t>0$ for all $t$ almost surely.

$\textcolor{green}{\textbf{Solution}}$

By Itô's lemma for a stochastic process $\hspace{2mm} X_t \hspace{2mm}$ satisfying $\hspace{2mm} dX_t = \mu(t, X_t)\,dt + \sigma(t, X_t)\,dW_t,\hspace{2mm}$ and a function $f(t, x) \in C^{1, 2}$, we have
$$
df(t, X_t) = \left(f_t^\prime + \mu(t, X_t)f_x^\prime + \frac{1}{2}\sigma^2(t, X_t)f_{xx}^{\prime\prime}\right)dt + \sigma(t, X_t)f_x^\prime\,dW_t
$$

Let $Y_t = \text{ln}(X_t). \newline$
For $f(x) = \text{ln}(x), \hspace{2mm} \mu(t, X_t) = a(t)X_t \hspace{2mm}$ and $\hspace{2mm} \sigma(t, X_t) = b(t)X_t, \hspace{2mm}$ we have
$$
\begin{align*}
    dY_t &= \left(0 + a(t)X_t\cdot \frac{1}{X_t} + \frac{1}{2}b^2(t)X_t^2\cdot \left(-\frac{1}{X_t^2}\right) \right)dt + b(t)X_t\cdot \frac{1}{X_t}\,dW_t \\
    &= \left(a(t)-\frac{1}{2}b^2(t)\right)dt + b(t)\,dW_t \\
\end{align*}
$$

Integrate:
$$
Y_t = Y_0 + \int_0^t{\left(a(s)-\frac{1}{2}b^2(s)\right)ds} + \int_0^t{b(s)\,dW_s}
$$

Exponentiate:
$$
X_t = X_0 \cdot \text{exp}\left(\int_0^t{\left(a(s)-\frac{1}{2}b^2(s)\right)ds} + \int_0^t{b(s)\,dW_s}\right)
$$

An exponential of any real number is positive with probability 1 and $X_0$ is positive by definition, thus $\hspace{1mm} \forall t \in \mathbb{R} \hspace{2mm} X_t > 0 \hspace{1mm}$ almost surely. \
Moreover, from $\hspace{1mm} dX_t = X_t \cdot (a(t)\,dt + b(t)\,dW_t) \hspace{1mm}$ it can be seen that the change in X is proportional to its value, therefore the solution is of exponential form...

$\textcolor{lightblue}{\textbf{Q2. (1.5 points)}}$ 

Derive the **Black–Scholes Partial Differential Equation (PDE)** starting from  
the dynamics  

$$
dS_t = \mu S_t\,dt + \sigma S_t\,dW_t, \qquad dB_t = rB_t\,dt.
$$  

(a) Define a self-financing portfolio $\Pi_t = V(t,S_t) - H_t S_t + G_t$.  
(b) Eliminate randomness to make $\Pi_t$ risk-free.  
(c) Show that under no-arbitrage,  

$$
V_t + \tfrac{1}{2}\sigma^2 S^2 V_{SS} + r S V_S - rV = 0.
$$  

(d) Briefly describe the financial meaning of each term.

$\textcolor{green}{\textbf{Solution}}$

Let $\hspace{1mm} V(t, S_t) \hspace{1mm}$ be the price of an option at time $\hspace{1mm} t$.\
Let $\hspace{1mm} B_t \hspace{1mm}$ be the price process of a risk-free asset with interest rate $\hspace{1mm} r, \hspace{1mm}$ following $\hspace{1mm} dB_t = rB_t\,dt$.\
The value at time $t$ of the portfolio consisting of a derivative, $H_t$ number of short-sold shares with price $S_t$, and $G_t$ of cash deposited under $r$:
$$
\Pi_t = V(t, S_t) - H_t S_t + G_t
$$
Self-financing condition for the portfolio — changes in holdings are financed internally by reallocating between its components with no external cash flow, which can be expressed as
$$
d\Pi_t = dV(t, S_t) - H_t\,dS_t + rG_t\,dt
$$
By Itô's lemma
$$
\begin{align*}
    dV(t, S_t) &= \left(
        V_t^\prime + \mu(t, S_t)V_s^\prime + \frac{1}{2}\sigma^2(t, S_t)V_{ss}^{\prime\prime}
    \right)dt +
        \sigma(t, S_t)V_s^\prime\,dW_t \\
    &= \left(
        V_t^\prime + \mu S_t V_s^\prime + \frac{1}{2}\sigma^2 S_t^2 V_{ss}^{\prime\prime}
    \right)dt +
        \sigma S_t V_s^\prime\,dW_t
\end{align*}
$$
Substitute $dV(t, S_t)$ and $dS_t$ to self-financing condition:
$$
\begin{align*}
    d\Pi_t &= \left(
        \left(V_t^\prime + \mu S_t V_s^\prime + \frac{1}{2}\sigma^2 S_t^2 V_{ss}^{\prime\prime}\right)dt + \sigma S_t V_s^\prime\,dW_t
    \right) - H_t\,\left(
        \mu S_t\,dt + \sigma S_t\,dW_t
    \right) + rG_t\,dt \\
    &= \left(
        V_t^\prime + \mu S_t V_s^\prime + \frac{1}{2}\sigma^2 S_t^2 V_{ss}^{\prime\prime} -
        \mu H_t S_t +
        rG_t
    \right)dt + \left(
        \sigma S_t V_s^\prime -
        \sigma H_t S_t 
    \right)dW_t \\
    &= \left(
        V_t^\prime + \frac{1}{2}\sigma^2 S_t^2 V_{ss}^{\prime\prime}
        + \mu S_t(V_s^\prime - H_t) +
        rG_t
    \right)dt +
        \sigma S_t (V_s^\prime - H_t)\,dW_t \\
\end{align*}
$$
Eliminate randomness by choosing the hedge $\hspace{1mm} H_t = V_s^{\prime}(t, S_t)$:
$$
d\Pi_t = \left(
    V_t^\prime + \frac{1}{2}\sigma^2 S_t^2 V_{ss}^{\prime\prime} + rG_t
\right)dt
$$
Now random shocks are compensated by the hedge.\
Riskless portfolio must earn the risk-free rate, otherwise arbitrage exists. That is $\hspace{1mm} d\Pi_t = r\Pi_t\,dt, \hspace{1mm}$ so
$$
\left(
    V_t^\prime + \frac{1}{2}\sigma^2 S_t^2 V_{ss}^{\prime\prime} + rG_t
\right)dt
=
r\left(
    V - H_t S_t + G_t
\right)dt
$$
Considering $\hspace{1mm} H_t = V_s^{\prime}(t, S_t), \hspace{1mm}$ this is
$$
V_t^\prime + \frac{1}{2}\sigma^2 S_t^2 V_{ss}^{\prime\prime} + rG_t
=
r\left(
    V - V_s^{\prime} S_t + G_t
\right)
$$
Which simplifies to
$$
V_t^\prime + \frac{1}{2}\sigma^2 S_t^2 V_{ss}^{\prime\prime} + rV_s^{\prime}S_t - rV = 0
$$
Financial interpretation:

$V_t^\prime \hspace{2mm}$ — time effect. How option price changes as time passes.\
$\frac{1}{2}\sigma^2 S_t^2 V_{ss}^{\prime\prime} \hspace{2mm}$ — the contribution of volatility and curvature.\
$rV_s^{\prime}S_t \hspace{2mm}$ — expected growth from holding the underlying asset.\
$- rV \hspace{2mm}$ — discounting at the risk-free rate.

$\textcolor{lightblue}{\textbf{Q3. (1.5 points)}}$  

(a) Explain the role of the **risk-neutral measure** $Q$.  
(b) Using Girsanov’s theorem, show how to transform  

$$
dS_t = \mu S_t dt + \sigma S_t dW_t
$$  

into  

$$
dS_t = r S_t dt + \sigma S_t dW_t^Q.
$$  

(c) Why does $\mu$ disappear from pricing formulas under $Q$?

$\textcolor{green}{\textbf{Solution}}$

**(a)** Define money market as $\hspace{1mm} dB_t = rB_t\,dt$.\
$Q$ is an equivalent probability measure, under which discounted asset prices $\hspace{1mm} \tilde{S_t} = \frac{S_t}{B_t} = e^{-rt}S_t \hspace{1mm}$ are martingales, that is $\hspace{1mm} \mathbb{E}^Q[\tilde{S_t} | \mathcal{F_t}] = \tilde{S}_{t-1}$. It moves investors' risk preferences out of scope and ensures all assets (including the particular one) earn the risk-free rate instead of the uncertain drift, allowing to compute unique, arbitrage-free derivative prices.

**(b)** Our goal is to achieve zero drift in $\hspace{1mm} \tilde{S_t}$.
$$
\begin{align*}
    d\tilde{S_t} = d(e^{-rt}S_t) &= e^{-rt}dS_t + S_td(e^{-rt}) + d(e^{-rt})dS_t \\
    &= e^{-rt}dS_t + S_td(e^{-rt}) \\
    &= e^{-rt}\left((\mu S_t dt + \sigma S_t dW_t) - r S_t d_t\right) \\
    &= e^{-rt}\left((\mu-r) S_t dt + \sigma S_t dW_t\right)
\end{align*}
$$
$\tilde{S_t} \hspace{1mm}$ has drift $\hspace{1mm} \mu-r \neq 0,$ so we introduce the market price of risk $\hspace{1mm} \theta = \frac{\mu - r}{\sigma}$.

By Girsanov Theorem $\hspace{1mm} \exists Q \sim P \hspace{1mm}$, such that the process
$$
W_t^Q = W_t + \int_0^t{\theta ds} = W_t + \theta t
$$
is a Brownian motion under $Q$.\
Then,

$$
\begin{align*}
    dS_t &= \mu S_t dt + \sigma S_t dW_t \\
    &= \mu S_t dt + \sigma S_t \left(dW_t^Q - \theta dt\right) \\
    &= (\mu - \sigma\theta)S_t\,dt + \sigma S_t\,dW_t^Q \\
    &= r S_t\,dt + \sigma S_t\,dW_t^Q \\
\end{align*}
$$
So that $\hspace{1mm} \tilde{S_t} \hspace{1mm}$ is a martingale under $Q$:
$$
\begin{align*}
    d\tilde{S_t} = d(e^{-rt}S_t) &= e^{-rt}\,dS_t + S_t\,d(e^{-rt}) \\
    &= e^{-rt}\left(r S_t\,dt + \sigma S_t\,dW_t^Q\right) + S_t \cdot r (e^{-rt})dt \\
    &= e^{-rt}\left(r S_t\,dt + \sigma S_t\,dW_t^Q - r S_t\,dt\right) \\
    &= e^{-rt}\sigma S_t\,dW_t^Q
\end{align*}
$$

**(c)**
It was shown explicitly in (b).\
$\mu$ is the compensation investors require for the risk under natural measure $P$. But under $Q$ we have hypothetical risk-neutral probabilities, where investors are indifferent to risk — so the compensation disappears and the drift becomes $r$. Hence pricing formulas depend on $r$ and $\sigma$, not on $\mu$

# **Block 2: Simulations**

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.stats as st
import yfinance as yf

np.random.seed(42)
plt.rcParams["figure.figsize"] = (10, 6)
```

```python
# Standard Brownian motion
def bm(t, n, x0=0, mu=0, sigma=1):
    dt = t/n
    return np.cumsum(np.concatenate([[0], np.random.randn(n)*sigma*np.sqrt(dt) + mu*dt]))
```

```python
# Geometric Brownian motion
def gbm(t, n, x0=1, mu=0, sigma=1):
    return x0*np.exp(bm(t, n, 0, mu-sigma**2/2, sigma))
```

$\textcolor{lightblue}{\textbf{Q4. (2 points)}}$

In this task, you will use **real stock market data** to explore whether
the **Geometric Brownian Motion (GBM)** model can approximate real asset dynamics.

Use **any publicly traded company** (for example: MSFT, NVDA, TSLA, META, or your choice).  

(a) Download **1–3 years** of daily price data using `yfinance`.  
(b) Compute **daily log-returns**:

r_t = \ln\frac{S_t}{S_{t-1}}
$$  
(c) Using these estimates, simulate a GBM path over the same time period.  
(d) Plot **actual vs simulated price paths** on the same chart.  
(e) Discuss: does GBM appear to capture the qualitative behavior of the real stock?

$\textcolor{green}{\textbf{Solution}}$

```python
# Of your choice
TICKER: str = 'AAPL'

# Close price reflects the consensus value at the end of trading, after all intraday volatility
# and it's a common choice, so we take it
price = yf.download(TICKER, start='2023-01-01', end='2025-11-03', interval='1d')['Close']
price.columns = ['Price']
price = price['Price']

# mu and sigma of log-returns correspond to parameters of GBM by construction
log_returns = np.log(price / price.shift(1)).dropna()
mu = log_returns.mean() * 252             # Annualized mean
sigma = log_returns.std() * np.sqrt(252)  # Annualized volatility

T = len(price) / 252  # Time in years
t = np.linspace(0, T, len(price))
motion = gbm(T, len(price)-1, x0=price.iloc[0], mu=mu, sigma=sigma)

plt.plot(t, price, 'r-', label=TICKER)
plt.plot(t, motion, 'b-', label='GBM Simulation')
plt.title(f'{TICKER} Stock Price vs GBM Simulation over {len(price)} Trading Days', fontsize=12, fontweight='bold')
plt.xlabel('Years', fontsize=12)
plt.ylabel('Price', fontsize=12)
plt.grid(True)
plt.legend()
plt.show()
```

Visual comparison of price paths is not significant enough to make judgments on how well GMB approximates real market data.\
Let us simulate large number of GBM paths and calculate daily Log-RMSE and Log-MAE between the simulations and the actual price, that is
$$
\text{Log-RMSE} = \sqrt{\text{Log-MSE}} = \sqrt{\frac{1}{n} \sum_t{(\text{ln}(S_t^{\text{sim}}) - \text{ln}(S_t^{\text{real}}))^2}}
$$
$$\text{and}$$
$$
\text{Log-MAE} = \frac{1}{n} \sum_t{|\text{ln}(S_t^{\text{sim}}) - \text{ln}(S_t^{\text{real}})|} 
$$
Then compute their means and evaluate typical price deviation of GBM simulation from the stock price, that we can use to assess GBM's qualitative fit. 

```python
M = 5000

rmse_log = np.zeros(M)
mae_log = np.zeros(M)
for i in range(M):
    path = gbm(T, len(price)-1, x0=price.iloc[0], mu=mu, sigma=sigma)
    rmse_log[i] = np.sqrt(np.mean((np.log(price) - np.log(path))**2))
    mae_log[i] = np.mean(np.abs(np.log(price) - np.log(path)))

plt.hist(mae_log, bins=50, alpha=0.5, color='green', label='Log-MAE')
plt.hist(rmse_log, bins=50, alpha=0.5, color='blue', label='Log-RMSE')
plt.title('Log-RMSE and Log-MAE Distribution over GBM Simulations', fontsize=12, fontweight='bold')
plt.grid(True)
plt.legend()
plt.show()

rmse_log_mean = rmse_log.mean()
mae_log_mean = mae_log.mean()
print(f'Log-RMSE: Mean = {np.around(rmse_log_mean, decimals=4)}, Std = {np.around(rmse_log.std(), decimals=4)}')
print(f'Log-MAE: Mean = {np.around(mae_log_mean, decimals=4)}, Std = {np.around(mae_log.std(), decimals=4)}')
```

Log-RMSE is actually expected to be higher as it penalizes greater deviations more.

The Typical Price Deviation can be found by

$$
\text{TPD } = \frac{S_t^{\text{sim}} - S_t^{\text{real}}}{S_t^{\text{real}}} = \frac{S_t^{\text{sim}}}{S_t^{\text{real}}} - 1 = e^{\text{ln}\left(\frac{S_t^{\text{sim}}}{S_t^{\text{real}}}\right)} - 1 = e^{\text{ln}\left(S_t^{\text{sim}}\right) - \text{ln}\left(S_t^{\text{real}}\right)} - 1 = e^{e_t} - 1
$$
where $e_t$ is the error term we have just estimated, thus we conclude that

Judging by RMSE, on average GBM paths deviate from the true stock by roughly $\quad e^{\text{Log-RMSE Mean}} - 1$\
Judging by MAE, on average GBM paths deviate from the true stock by roughly $\quad e^{\text{Log-MAE Mean}} - 1$


```python
print(f'Typical multiplicative deviation = {np.around(np.exp(rmse_log_mean) - 1, decimals=4) * 100}%')
print(f'Average absolute deviation = {np.around(np.exp(mae_log_mean) - 1, decimals=4) * 100}%')
```

That is closer to poor qualitative fit and I wound not call it satisfying.\
In conclusion, GBM does not appear to capture the qualitative behavior of the real stock efficiently.

$\textcolor{lightblue}{\textbf{Q5. (1.5 points)}}$

Simulate **Geometric Brownian Motion (GBM)** paths solving  

$$
dS_t = \mu S_t\,dt + \sigma S_t\,dW_t, \quad S_0 = 100.
$$  

Use parameters  
$$
\mu = 0.08, \quad \sigma = 0.2, \quad T = 1, \quad N = 252.

(a) Implement the Euler–Maruyama scheme.  
(b) Plot **10 sample paths** of $S_t$.  
(c) Compute the empirical mean and variance of $S_T$, and compare with the theoretical values:  

$$
E[S_T] = S_0 e^{\mu T}, \quad Var(S_T) = S_0^2 e^{2\mu T}(e^{\sigma^2 T} - 1).
$$

$\textcolor{green}{\textbf{Solution}}$

**(a)** For a SDE
$$
dX_t = \mu (t, X_t)\,dt + \sigma (t, X_t)\,dW_t
$$
Euler–Maruyama scheme is
$$
X_{t_{i+1}} = X_{t_i} + \mu (t_i, X_{t_i})\Delta t + \sigma (t_i, X_{t_i})\Delta W_i
$$
Since $\hspace{1mm} \Delta W_i \sim N(0, \Delta t) \hspace{1mm}$ we have $\hspace{1mm} \Delta W_i = \sqrt{\Delta t}\cdot Z, \hspace{1mm} Z \sim N(0, 1). \hspace{1mm}$\
Thus,
$$
X_{t_{i+1}} = X_{t_i} + \mu (t_i, X_{t_i})\Delta t + \sigma (t_i, X_{t_i}) Z_{t_i} \sqrt{\Delta t}, \quad Z \sim N(0, 1)
$$
In our case for now
$$
X_{t_{i+1}} = X_{t_i} + \mu (X_{t_i})\Delta t + \sigma (X_{t_i}) Z_{t_i} \sqrt{\Delta t}, \quad Z \sim N(0, 1)
$$

```python
def euler_maruyama(mu, sigma, T=1, N=1000, X0=1):
    dt = T/N
    X = np.zeros(N+1)
    X[0] = X0
    for i in range(N):
        X[i+1] = X[i] + mu(X[i])*dt + sigma(X[i])*np.random.randn()*np.sqrt(dt)
    return np.linspace(0, T, N+1), X
```

```python
mu = lambda X: 0.08 * X
sigma = lambda X: 0.2 * X
t, X = euler_maruyama(mu, sigma, T=1, N=252, X0=100)

plt.plot(t, X, 'g-')
plt.title('GBM Simulation via Euler-Maruyama', fontsize=12, fontweight='bold')
plt.xlabel('Time t', fontsize=12)
plt.ylabel('$S_t$', fontsize=12)
plt.grid(True)
plt.show()
```

**(b)** Generate 10 sample paths of $S_t$ using Euler-Maruyama scheme:

```python
M_em = 10

em_terminal = np.zeros(M_em)
for i in range(M_em):
    t, X = euler_maruyama(mu, sigma, T=1, N=252, X0=100)
    em_terminal[i] = X[-1]
    plt.plot(t, X, alpha=0.7)

plt.title(f'{M_em} GBM Simulations via Euler-Maruyama', fontsize=12, fontweight='bold')
plt.xlabel('Time t', fontsize=12)
plt.ylabel('$S_t$', fontsize=12)
plt.grid(True)
plt.show()
```

**(c)** Compute the empirical mean and variance of $S_T$ and compare with the theoretical values:

```python
emp_mean_em = np.mean(em_terminal)
emp_var_em = np.var(em_terminal)

# TODO: Introduce hyperparameters
theoretical_mean_gbm = 100 * np.exp(0.08*1)
theoretical_var_gbm = 100**2 * np.exp(2*0.08*1) * (np.exp(0.2**2*1)-1)

print(f'Empirical mean = {emp_mean_em}')
print(f'Theoretical mean = {theoretical_mean_gbm}')
print('---------')
print(f'Empirical variance = {emp_var_em}')
print(f'Theoretical variance = {theoretical_var_gbm}')
```

$\textcolor{lightblue}{\textbf{Q6. (1.5 points)}}$  

Verify that under the risk-neutral measure, the **discounted stock price**  

$$
\tilde{S}_t = e^{-rt} S_t
$$  

is a **martingale**, i.e.  
$$
E_Q[\tilde{S}_T \mid \mathcal{F}_t] = \tilde{S}_t.
$$  

(a) Simulate $10{,}000$ GBM paths under $Q$.  
(b) Plot the average of $\tilde{S}_t$ over time and compare with the initial value $S_0$.  
(c) Comment on the result.

$\textcolor{green}{\textbf{Solution}}$

```python
T_q = 1.0
n = 252
t = np.linspace(0, T_q, n+1)
r = 0.04
S0 = 100

M_q = 10_000
paths_q = list()
for _ in range(M_q):
    paths_q.append(gbm(T_q, n, x0=S0, mu=r, sigma=0.2))
paths_q = np.stack(paths_q, axis=1) # 2D Array of shape (n+1) x M_q
                                    # Now paths are stored column-wise
                                    # Each row corresponds to paths’ values at the same t
B = np.exp(r*t) * S0 # Risk-free money market

# Plot couple of paths comparing to B_t
for i in range(5):
    S = paths_q[:, i]
    plt.plot(t, S, alpha=0.7)
plt.plot(t, B, 'black', label='$B_t$')
del S
plt.title(f'[GBM Paths under $Q$] vs $B_t$', fontsize=12, fontweight='bold')
plt.xlabel('Time t', fontsize=12)
plt.ylabel('$S_t$', fontsize=12)
plt.grid(True)
plt.legend()
plt.show()
```

Now we are about to verify $\hspace{1mm} \mathbb{E}^Q[\tilde{S}_t] \approx \tilde{S}_0 = S_0$.

```python
disc_factor = np.exp(-r*t)[:, None]
S_tilde = paths_q * disc_factor # [(n+1) x M_q] by [(n+1) x 1] ->
                                # Each row (timestamp) is multiplied by the corresponding disc_factor
S_tilde_t_mean = S_tilde.mean(axis=1) # The mean of disc prices across paths for each t

plt.plot(t, S_tilde_t_mean, 'b-')
plt.axhline(S0, color='r', linestyle='--')
plt.title(r'$\mathbb{E}^Q[\tilde{S}_t] \approx S_0 = \text{const}$', fontsize=12)
plt.xlabel('Time t', fontsize=12)
plt.ylabel('Mean discounted price', fontsize=12)
plt.grid(True)
plt.legend()
plt.show()
```

Indeed, the mean discounted price over timestamps t appeared to be near constant (that is, it does have no trend), indicating $\hspace{1mm} \tilde{S} \hspace{1mm}$ is a $Q$-martingale by definition.
